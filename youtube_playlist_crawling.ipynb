```python
# url 주소 입력
url = input('>>>> URL 주소소를 입력 : ')
```

```python

# https://www.youtube.com/watch?v=7RtWlPpk348&list=PLwvr-xPygMX9TaQFW3C1UGEuD0zJF7pCk

# # 저장할 파일명
# save_name = input('>>> 저장할 파일명 이름을 입력 : ') + '.xlsx'

# 제목에서 제거할 텍스트
replace_text  = input('>> 제거할 텍스트를 입력하세요 : ')

# 웹브라우저 작동을 위한 라이브러리
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
import urllib.request

# HTML Parsing을 위한 라이브러리
from bs4 import BeautifulSoup

# 웹브라우저 작동을 기다리기 위한 라이브러리
import time
import random

# IPython
from IPython.display import display

# 경고 무시
import warnings
warnings.filterwarnings(action='ignore')

# 데이터프레임 및 CSV 파일 저장을 위한 라이브러리
import pandas as pd

# 시간 관련 라이브러리
from datetime import datetime, timedelta
from pytz import timezone

# 데이터프레임 출력
from tabulate import tabulate

# User-Agent를 설정
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.134 Safari/537.36'

# 크롬드라이버 option설정
options = webdriver.ChromeOptions()
options.add_argument('user-agent='+ user_agent)

# 브라우저 생성
driver = webdriver.Chrome('C:/chromedriver.exe', options = options)

# 웹사이트 열기
driver.get(url)

# 로딩이 끝날 때까지 3초 정도 기다림
driver.implicitly_wait(3)

# Chrome 창 최대화
driver.maximize_window()
time.sleep(3)

# # 웹 브라우저의 HTML 정보 가져오기
# soup = BeautifulSoup(driver.page_source, 'lxml')

# 페이지 소스 추출
html_source = driver.page_source
soup_source = BeautifulSoup(html_source, 'lxml')

# Chrome 창 닫기
driver.close()

# 파싱정보 가져오기
parsing = soup_source.find_all('a', class_ = 'yt-simple-endpoint style-scope ytd-playlist-video-renderer')
video_time = soup_source.find_all('span', class_ ='style-scope ytd-thumbnail-overlay-time-status-renderer') #검색했을 때 검색숫자가 안맞아서 확인이 필요함  

# 파싱정보 정리하기
name_list = []
url_list = []
time_list = []

for i in range(len(parsing)):
    name_list.append(parsing[i].text.strip())
    main = 'https://www.youtube.com'
    sub = parsing[i].get('href')  
    url_list.append(f'{main}{sub}')
    time_list.append(video_time[i].text.strip())

    
playlist = {
    'title' : name_list,
    'No.' : range(1, len(parsing)+1),
    '시간' : time_list,
    '확인' : '',    
    'URL' : url_list, 
}

playlist = pd.DataFrame(playlist)
playlist['title'] = playlist['title'].apply(lambda x: x.replace(replace_text,'').strip())
playlist.style.hide_index()    

    
# # 파싱정보 데이터프레임 만들기    
# playlist_dic = {
#     'No.' : range(1, len(parsing)+1),
#     '제목' : name_list,
#     '시간' : time_list,
#     '확인' : '',    
#     'URL' : url_list, 
# }

# playlist_df = pd.DataFrame(playlist_dic)    
# playlist_df['제목'] = playlist_df['제목'].apply(lambda x: x.replace(replace_text,'').strip())


# # # # 파싱정보 저장하기
# # playlist_df.to_excel(save_name, encoding='CP-949', index=False)

# # 테스트 결과 출력하기
# print()
# print('================== Test 출력 ==================')
# print()

# print('출력 >> ' , playlist_df['제목'][0])
# print('출력 >> ' , playlist_df['URL'][0])
# print('출력 >> ' , playlist_df['시간'][0])

# print()
# print('================== Test 출력 ==================')

# print()
# print('>>> 저장이 완료되었습니다.')
# print()
```
